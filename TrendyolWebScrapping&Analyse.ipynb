```python
brands=[]
for wrap in wrapper:
    categories = wrap.find_all("div", class_="fltrs-wrppr hide-fltrs")
    for category in categories:
        info = category.find("div", class_="fltr-cntnr-ttl")
        if info and "Marka" in info.text.strip():
            print(info.text.strip())
            print("\n")
            fltr_items = category.find_all("div", class_="fltr-item-text")
            for item in fltr_items:
                print(item.text.strip())
                brands.append(item.text.strip())
```

In order to select the most preferred brands in the market and display link extensions to reach those pages directly.

```python
brands_url=[]
for wrap in wrapper:
    categories = wrap.find_all("div", class_="fltrs-wrppr hide-fltrs")
    for category in categories:
        info = category.find("div", class_="fltr-cntnr-ttl")
        if info and info.text.strip() == "Marka":
            fltr_items = category.find_all("a", class_="fltr-item-wrppr")
            for item in fltr_items:
                link = item.get("href")
                print(link)
                brands_url.append(link)
```

This function shows us the scoring according to the categories divided into percentage values with the width value in the HTML.

```python
def calculate_starts(rating_container):
    # Initialize total value
    total_value = 0
    for star in rating_container:
        width_style = star.get('style')
        width_percentage= float(width_style.split(";")[0].split(":")[1].strip().strip("%")) / 100
        if width_percentage <= 0.25:
            total_value += 0.25
        elif width_percentage <= 0.5:
            total_value += 0.5
        elif width_percentage <= 0.75:
            total_value += 0.75
        else:
            total_value += 1

    return total_value
```

It will be more useful to have an overview of the products before examining them in detail.

```python
product_page_detail=[]
brands_url_info = [
   '/apple-cep-telefonu-x-b101470-c103498',   #We will examine the most popular and competitive brands in the phone industry
   '/samsung-cep-telefonu-x-b794-c103498',
   '/xiaomi-cep-telefonu-x-b101939-c103498',
   '/huawei-cep-telefonu-x-b103505-c103498'
]

#Definition of base URL and page range for web scraping
base_url = "https://www.trendyol.com"
start_page = 1
end_page = 10

#Empty list to store the scraped data
data = []

# Data Collection
for brand_url in brands_url_info:
  for page_number in range(start_page, end_page + 1):

      url = f"{base_url}{brand_url}?pi={page_number}"
      url_phone_r=requests.get(url)
      soup = bs(url_phone_r.text,"lxml")

      products = soup.find_all("div",class_="p-card-wrppr with-campaign-view")

      for product in products:
        link_prd_page = product.find("div", class_="p-card-chldrn-cntnr card-border").find("a")["href"]
        page_link=f"{base_url}{link_prd_page}" #products main page link

        info_down = product.find("div",class_="product-down") #In the section of the wrapper divided into 2 section image and down
        if info_down:

          prd_desc_cntr  = info_down.find("div",class_="prdct-desc-cntnr")                  #product description container
          brand          = prd_desc_cntr.find("span",class_="prdct-desc-cntnr-ttl")         #brand name
          describe       = prd_desc_cntr.find("span",class_="prdct-desc-cntnr-name hasRatings") #brand description
          prd_price_cntr = info_down.find("div",class_="price-promotion-container")          #product price container
          price          = prd_price_cntr.find("div",class_="prc-box-dscntd")

          prd_rating_cntr = info_down.find("div", class_="ratings-container")       #product rating container
          if  prd_rating_cntr:
            filled_stars    = prd_rating_cntr.find_all("div", class_="full")         #getting stars width
            total_value     = calculate_starts(filled_stars)                          #calculate the rate
            rating_count    = prd_rating_cntr.find("span",class_="ratingCount")      #rating count depens on rating
            if rating_count:
                rating_count = rating_count.text.replace("(", "").replace(")", "")
            else:
                rating_count=0

          prd_prc_prom_cntr = info_down.find("div", class_="price-promotion-container") #product promotion container
          prom = prd_prc_prom_cntr.find("span",class_="low-price-title") #product price promotion
          if prom:
            prom=prom.text
          else:
            prom=""

        info_container = soup.find("div", class_="image-container") #product shippment information container
        if info_container:
            product_stamps = info_container.find("div", class_="product-stamps-container")
            if product_stamps:
              if product_stamps.find("div", class_="product-stamp-container freeCargo"):#product free delivery
                free_delivery="exist"
              else:
                free_delivery=""
              if product_stamps.find("div", class_="product-stamp-container rushDelivery"):#product fast delivery
                fast_delivery="exist"
              else:
                fast_delivery=""
            else:
                print("Product stamp information could not be found")

        product_page_detail.append(page_link);
        products_data = {
            "Link":  page_link,
            "Brand": brand.text,
            "Original Price": price.text,
            "Rate":total_value,
            "RateCounts":rating_count,
            "Free delivery":free_delivery,
            "Fast Delivery":fast_delivery,
            "Promotion":prom

          }
        data.append(products_data)

#Convert dataset to a DataFrame
df = pd.DataFrame(data)
print(df)

#Converting it to CSV so we can download
df.to_csv('trendyol_all_data.csv', encoding="utf-8")

```

We collect product page information in a single format

```python
all_details = []
url_phone=product_page_detail


for url in url_phone:
    url_phone_r = requests.get(url)
    soup = bs(url_phone_r.text, "lxml")

    prd_wrp = soup.find("div", class_="product-detail-wrapper")  # product wrapper
    if  prd_wrp:
      prd = prd_wrp.find("div", class_="pr-in-w")  # prd information
      if prd:
          details = {}

          detail = prd.find("h1", class_="pr-new-br").find('span')  # details
          details["Details"] = detail.text

          brand = prd.find("a", class_="product-brand-name-with-link")  # brand
          details["Brand"] = brand.text

          price_cntr = prd.find("div", class_="product-price-container")
          price = price_cntr.find("span", class_="prc-dsc")
          details["Price"] = price.text

          detail_items = soup.find_all("li", class_="detail-attr-item")
          for item in detail_items:
              attribute = item.find("span").text.strip()
              value = item.find("b").text.strip()
              details[attribute] = value

          all_details.append(details)

# Create DataFrame
df = pd.DataFrame(all_details)
print(df)
```
