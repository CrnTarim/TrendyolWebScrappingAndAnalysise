# Brand Rate - Number of Votes Relationship Pie Chart Diagram

We display the brand distributions of the products that have received 20 or more votes in terms of reliability between 0-4.00, 4.00-4.25 4.25-5.50 4.50-4.75 4.75-5.

```python
rates0_4 =df_all[(df_all.Rate >= 0) & (df_all.Rate < 4.00) & (df_all.RateCounts >=20)].groupby(["Brand"]).size().reset_index(name='Count').sort_values(by='Count', ascending=False).head()
rates25_50=df_all[(df_all.Rate >= 4.00) & (df_all.Rate < 4.25) & (df_all.RateCounts >=20)].groupby(["Brand"]).size().reset_index(name='Count').sort_values(by='Count', ascending=False).head()
rates50_75=df_all[(df_all.Rate >= 4.25) & (df_all.Rate < 4.50) & (df_all.RateCounts >=20)].groupby(["Brand"]).size().reset_index(name='Count').sort_values(by='Count', ascending=False).head()
rates75_0=df_all[(df_all.Rate >= 4.50) & (df_all.Rate < 4.75) & (df_all.RateCounts >=20)].groupby(["Brand"]).size().reset_index(name='Count').sort_values(by='Count', ascending=False).head()
rates75_5=df_all[(df_all.Rate >= 4.75) & (df_all.Rate <= 5.00) & (df_all.RateCounts >=20)].groupby(["Brand"]).size().reset_index(name='Count').sort_values(by='Count', ascending=False).head()
```

```python
def create_pie_chart(data, title):
    fig = px.pie(data, values = 'Count', names = 'Brand', title=title)
    fig.show()

create_pie_chart(rates0_4, '0-4 Ratings by Brand')
create_pie_chart(rates25_50, '4-4.25 Ratings by Brand')
create_pie_chart(rates50_75, '4.25-4.5 Ratings by Brand')
create_pie_chart(rates75_0, '4.5-4.75 Ratings by Brand')
create_pie_chart(rates75_5, '4.75-5 Ratings by Brand')
```

# Analysis of Brands According to Key Features

```python
avg_price = df_all.groupby('Brand')['Original Price'].mean().reset_index(name='AVG').sort_values(by='AVG', ascending=False)
print(avg_price)
```

## Avarage Price

```python
diverging_colors = sns.color_palette("RdBu", 10)
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price, x='Brand', y='AVG', hue='Brand', palette=diverging_colors, legend=False)
plt.title('Average Prices by Brands')
plt.xlabel('Brand')
plt.ylabel('AVG Price')
plt.xticks(rotation=45)
plt.show()
```

**We observe that the iPhone has a very high price by far, so let's evaluate the price performance of the brands.**

**The battery is indispensable for a phone and is also very important to the consumer.**

```python
brand_memory=df_main.groupby(["Brand","Dahili Hafıza"]).size().reset_index(name='Count').sort_values(by='Dahili Hafıza', ascending=False)
brand_memory.head(50)
```

## Internal Memory

```python
custom_palette = sns.color_palette("icefire", as_cmap=True)
plt.figure(figsize=(10, 6))
sns.barplot(data=brand_memory, x='Brand', y='Count', hue='Dahili Hafıza', palette=custom_palette)
plt.title('Brand and Internal Memory Amount')
plt.xlabel('Brand')
plt.ylabel('GB')
plt.xticks(rotation=45)
plt.legend(title='Internal Memory')
plt.show()
```

# Fast Delivery and Free Shipping Product in the Campaign

**According to the chart here, the selection of products on the left side of the chart is more affordable.**

```python
promotion_apple=df_all[df_all["Brand"] == "Apple"][["Brand","Original Price","Promotion", "Free delivery", "Fast Delivery"]].sort_values(by="Original Price", ascending=False).dropna()
promotion_apple
```

```python
promotion_samsung=df_all[df_all["Brand"] == "Samsung"][["Brand","Original Price","Promotion", "Free delivery", "Fast Delivery"]].sort_values(by="Original Price", ascending=False).dropna()
promotion_xiaomi=df_all[df_all["Brand"] == "Xiaomi"][["Brand","Original Price","Promotion", "Free delivery", "Fast Delivery"]].sort_values(by="Original Price", ascending=False).dropna()
promotion_huawei=df_all[df_all["Brand"] == "Huawei"][["Brand","Original Price","Promotion", "Free delivery", "Fast Delivery"]].sort_values(by="Original Price", ascending=False).dropna()

```

```python
promotion_samsung.head(1)
```

```python
all_promotions = pd.concat([promotion_apple, promotion_samsung, promotion_xiaomi])
all_promotions
```

##Promotion Period Distribution by Brands

```python
promo_counts = all_promotions.groupby(['Brand', 'Promotion']).size().unstack(fill_value=0).reset_index()

fig = px.bar(promo_counts.melt(id_vars='Brand', var_name='Promotion', value_name='Count'),
             x='Brand', y='Count', color='Promotion', barmode='stack',
             title='Promotion Period Distribution by Brands', labels={'Count': 'Number of Promotions'})
fig.update_layout(xaxis_title='Marka', yaxis_title='Number of Promotions', xaxis_tickangle=-45)
fig.show()
```

## The most expensive product in this category and its price

```python
apple_first_row = promotion_apple.head(1)
samsung_first_row = promotion_samsung.head(1)
xiaomi_first_row = promotion_xiaomi.head(1)

df = pd.concat([apple_first_row, samsung_first_row, xiaomi_first_row])

fig = px.pie(df, values='Original Price', names=df.Brand, title='The Most Expensive Product in the Campaign')
fig.show()
```
